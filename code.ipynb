```python
import pandas as pd
```

```python
df = pd.read_csv('attendance.csv' )
df
```

```python
file_path = 'today.txt'
with open(file_path, 'r') as file:
        content = file.read()
        numbers_list = [int(num.strip()) for num in content.split(',')]
        
for i, rollno in enumerate(df['Rollno']):
    if int(rollno[-2:]) in numbers_list[1:]:
        df.at[i, f'{numbers_list[0]} March'] = 0
df[f'{numbers_list[0]} March'] = df[f'{numbers_list[0]} March'].fillna(1).astype(int)

column_names = list(df.columns)
column_names[-2], column_names[-1] = column_names[-1], column_names[-2]
df = df[column_names]

df['Total_attendance'] = df[list(df.columns[2:-1])].sum(axis=1)        
```

```python
df.to_csv('attendance.csv',index=False)
```
